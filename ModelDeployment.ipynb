```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/salary_predict_dataset (2).csv')
df
```

```python
#convert string values to numerical value

df.isnull().sum()
```

```python
df.info
```

```python
#fill score with average value
df['test_score'].fillna(df['test_score'].mean(),inplace=True)
df
```

```python
#process string and replace by numerical value and replace NaN by null
df['experience'].fillna(0,inplace=True)
df
```

```python
def string_to_number(word):
  dict={'one':1,'three':3,'five':5,'six':6,'ten':10,'fifteen':15,'thirteen':13,'four':4,'eleven':11,'twelve':12,0:0}
  return dict[word]
```

```python
df['experience']=df['experience'].apply(lambda x: string_to_number(x))
df
```

```python
df['interview_score'].fillna(df['interview_score'].mean(),inplace=True)
df
```

```python
#training the model

x=df.iloc[:,:3]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.67,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
mymodel=LinearRegression()
mymodel.fit(x_train,y_train)
```

```python
y_pred=mymodel.predict(x_test)
```

```python
y=mymodel.predict([[4,5,7]])
# 4=experience,5=test_score,7=interview_score
y
```

```python
y=mymodel.predict([[10,9,10]])
# 10=experience,9=test_score,10=interview_score
y
```

```python
#we save the model using pickle
import pickle
pickle.dump(mymodel,open("model.pkl","wb")) #i want to save it in variable called model.pkl
```
