```python
import pandas as pd
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
dummies=pd.get_dummies(df.State)
dummies
```

```python
# concate dummie to original dataset

merge=pd.concat([df,dummies],axis='columns')
print(merge)
```

```python
# drop state column feom dataframe

final=merge.drop(['State'],axis='columns')
print(final)
```
