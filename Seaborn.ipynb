```python
import seaborn as sns
sns.__version__
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names
```

```python
tips=sns.load_dataset('tips')
tips.head()

```

```python
#simple bar plots

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',hue='size',data=tips)
```

```python
#palette attribute


sns.palplot(sns.color_palette('deep',5))
sns.palplot(sns.color_palette('spring',7))
```

# creating our color palettes

```python
my=['red','black','green','yellow','pink','violet']
sns.set_palette(my)

sns.palplot(sns.color_palette())
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#relational plot with scatter plot

sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips,style="time")
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips,style="time",size="size")
```
